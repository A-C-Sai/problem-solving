```python
'''
Properties of Asymptotic Notation
-----------------------------------

Genearal Properties
--------------------

if f(n) is O(g(n)) then a*f(n) is O(g(n))

Applies to Omega and Thetha also
 
Reflexive
----------

if f(n) is given then f(n) is O(f(n))

Applies to Omega also

Transitive
-----------

if f(n) is O(g(n)) and g(n) is O(h(n)) then f(n) is O(h(n))

Applies to Omega and Thetha also

Symmetric
----------

if f(n) is Thetha(g(n)) then g(n) is Thetha(f(n))

Only applies to Thetha 

Transpose Symmetric
--------------------

if f(n) is O(g(n)) then g(n) is Omega(f(n))

Applies to Big-Oh and Omega only


----------------------------------------------------------------------

if f(n) = O(g(n))
and f(n) = Omega(g(n))
f(n) = Thetha(g(n))

----------------------------------------------------------------------

if f(n) = O(g(n))
and d(n) = O(e(n))

then f(n) + d(n) =  O(Max(g(n),e(n)))

----------------------------------------------------------------------

if f(n) = O(g(n))
and d(n) = O(e(n))

then f(n)*d(n) =  O(g(n)*e(n))


'''
print()
```
